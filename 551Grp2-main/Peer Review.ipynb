# Project Peer Review

The purpose of this individual report is to reflect on the projects
conducted by other groups as an external evaluator. 
For this part, you should be present during the group presentations in the final week of the class. 

The peer-review is anonymous. It means your reviewer is not revealed to the group whose work is under review. 

To submit your peer review work, use the following criteria:

* Clarity of the presentation, organization, and narrative coherence

* Delivery criteria (voice, pacing, presence)

* Clarity of the visuzalitions (e.g., are the plots appropriate to answer selected research questions)

* Quality of the dashboard and visualizations including the quantity of the plots, choice of appropriate EDA criteria, and interactivity of the dashboard. See the project notebook for details. 



Your review should be constructive and help the groups know what they have done wrong and how to improve it. 

See some reviewing guidelines [here](https://guides.library.queensu.ca/journal-publishing-guide/peer-review)

Some notes from the above link to consider when you review:
-	Reviewers should spend the appropriate and necessary time to provide feedback
-	Avoid making comments or criticisms based on your own personal style. Refer to accepted guidelines, concepts and rules and present your suggestions fairly and clearly.
-	Fully explain the problems you find. Provide specific details from the draft to illustrate. Attempt to suggest ways to correct the problem. 
-	Find positive, encouraging things to say about the draft that you are reviewing. 

Remember to be constructive and write about issues you find, and suggest how to improve them. 
Your feedback and grading **will not** affect our grading!
