# EDA: Healthcare Survey dataset
This notebook downloads the Kaggle dataset and prepares it for exploration.

```

import kagglehub
from pathlib import Path

# Kaggle dataset slug. Update if this dataset is renamed or removed.
DATASET_HANDLE = 'aradhanahirapara/healthcare-survey'

# Download into a local folder within this milestone
DATA_DIR = Path('data/healthcare_survey')
DATA_DIR.mkdir(parents=True, exist_ok=True)

try:
    path = kagglehub.dataset_download(DATASET_HANDLE, path=str(DATA_DIR))
    print('Path to dataset files:', path)
    print('Contents:')
    for p in Path(path).iterdir():
        print('-', p.name)
except Exception as e:
    print(f"Download failed for {DATASET_HANDLE}: {e}")
    print('If the slug is incorrect or the dataset was removed, update DATASET_HANDLE to a valid Kaggle dataset.')
```
